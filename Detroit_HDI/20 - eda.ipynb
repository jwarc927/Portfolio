## Distributions of Human Development Indices in the USA

Before we progress to modeling it may be illustrative to examine the distribution of some of the indices we calculated.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('./data/clean_data.csv')
```

```python
df.info()
```

The chart below demonstrates that the HDIs are fairly normally distributed accross the counties in our dataset.  This is good to see because it suggests that the model may be able to extract some differentability.

```python
sns.color_palette("YlOrBr", as_cmap=True)
sns.violinplot(x = df['HDI'], color = "brown").set(title = "Density of HDI Values")
plt.savefig("./assets/hdi", transparent=True)
```

We also see that the Income Index (II) is normally distributed.

```python
sns.color_palette("YlOrBr", as_cmap=True)
sns.violinplot(x = df['II'], color = "brown").set(title = "Density of II Values")
plt.savefig("./assets/ii", transparent=True)
```

The Educational Index (EI) is also fairly normal, but note that it is a narrower distribution than income.  This suggests that years of education are fairly standardized in the United States with relatively few counties having a much larger proportion of highschool dropouts or advanced degree earners from the mean.

```python
sns.color_palette("YlOrBr", as_cmap=True)
sns.violinplot(x = df['EI'], color = "brown").set(title = "Density of EI Values")
plt.savefig("./assets/ei", transparent=True)
```

We do not see the same pattern in the Life Expectancy Index (LE). It appears that there are a large number of counties that see considerably lower life expectancies than the American Mean.  We may expect to see a lot of impact on health and wellness features in our model of the HDI.

```python
sns.color_palette("YlOrBr", as_cmap=True)
sns.violinplot(x = df['LE'], color = "brown").set(title = "Density of LI Values")
plt.savefig("./assets/le", transparent=True)
```

Finally, the large spread on the box plot below shows that there are many features that are highly correlated to the II.  This suggests that in the next step we will remove a lot of income related features to reduce colinearity in our features.

```python
sns.color_palette("YlOrBr", as_cmap=True)
sns.boxplot(x = df.corr()['II'], color = "brown").set(title = "Feature Correlation to II")
plt.savefig("./assets/iicorr", transparent=True)
```

We see a similar pattern in the correlations to EI, although here the spread is somewhat smaller, so we should not expect to need to remove as many educational features.

```python
sns.color_palette("YlOrBr", as_cmap=True)
sns.boxplot(x = df.corr()['EI'], color = "brown").set(title = "Feature Correlation to EI")
plt.savefig("./assets/eicorr", transparent=True)
```

Lastly, colinearity of features with respect to the LE is moderate.  Some of these features will likely be removed as well, but not to the extent of the incoome related features.

```python
sns.color_palette("YlOrBr", as_cmap=True)
sns.boxplot(x = df.corr()['LE'], color = "brown").set(title = "Feature Correlation to LE")
plt.savefig("./assets/lecorr", transparent=True)
```
